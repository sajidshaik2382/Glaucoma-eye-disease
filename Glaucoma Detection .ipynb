```python
object_columns = df.select_dtypes(include=['object']).columns
print("object type columns:")
print(object_columns)

numerical_columns = df.select_dtypes(include=['int64','float64']).columns
print("numerical type columns:")
print(numerical_columns)
```

```python
def classify_features(df):
    categorical_features = []
    non_categorical_features = []
    discrete_features =[]
    continous_features = []

    for column in df.columns:
        if df[column].dtype =='object':
            if df[column].nunique() < 10:
                categorical_features.append(column)
            else:
                non_categorical_features.append(column)
        elif df[column].dtype in ['int64','float64']:
                if df[column].nunique() < 10:
                    discrete_features.append(column)
    return categorical_features, non_categorical_features,discrete_features,continous_features
```

```python
categorical_features, non_categorical_features,discrete,continous = classify_features(df)
```

```python
print("categorical_features:", categorical_features)
print("Non-Categorical Features:",non_categorical_features)
```

```python
print("discrete_features",discrete)
print("continous_features:",continous)
```

```python
for i in categorical_features:
    print(i,':',df[i].unique())
    print()
```

```python
for i in categorical_features:
    plt.figure(figsize=(15,6))
    sns.countplot(x = df[i], data = df, palette = 'hls')
    plt.show()
```

```python
for i in categorical_features:
    plt.figure(figsize=(30,20))
    plt.pie(df[i].value_counts(), labels=df[i].value_counts().index,
            autopct='%1.1f%%', textprops={'fontsize' :25,
                                        'color':'black',
                                        'weight' : 'bold',
                                        'family':'serif'})
    hfont = {'fontname':'serif','weight':'bold'}
    plt.title(i,size = 20,**hfont)
    plt.show()
```
